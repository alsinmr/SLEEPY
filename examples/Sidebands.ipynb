## Define spin system

```python
ex=RS.ExpSys(v0H=600,Nucs=['15N','1H'],vr=0,pwdavg=RS.PowderAvg('zcw4180'))
ex.set_inter(Type='dipole',i0=0,i1=1,delta=20000)
ex.set_inter(Type='CSA',i=0,delta=100,eta=.5)

L=RS.Liouvillian((ex))
L.add_relax(Type='T2',i=0,T2=.01)
seq=RS.Sequence(L)
seq.add_channel('1H',t=[0,1/5000/10],v1=0)
rho=RS.Rho(rho0='15Nx',detect='15Np',L=L)
```

```python
rho.clear()
rho.DetProp(seq=seq,n=500)
```

```python
rho.plot(FT=True,apodize=True)
```

```python
rho.plot()
```
