```python
ex0=sl.ExpSys(v0H=600,Nucs='2H',LF=True,vr=0)
ex0.set_inter('quadrupole',i=0,delta=128000)
ex1=ex0.copy()
ex1.set_inter('quadrupole',i=0,delta=64000)
```

```python
L=sl.Liouvillian((ex0,ex1))
# L=ex0.Liouvillian()
L.kex=sl.Tools.twoSite_kex(tc=5e-5,p1=.5)
# seq=L.Sequence().add_channel('2H',t=[0,ex0.taur/20])
seq=L.Sequence().add_channel('2H',t=[0,1e-6])
```

```python
rho=sl.Rho(rho0='2Hx',detect='2Hp')
rho.DetProp(seq,n=200)
rho.downmix()
```

```python
rho.plot(FT=True,apodize=True,axis='kHz')
```
